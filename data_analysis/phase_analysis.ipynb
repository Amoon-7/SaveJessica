```python
import numpy as np
import matplotlib.pyplot as plt

# Example signal
fs = 1            # Sampling rate
t = np.arange(0, 1000)
f = 0.1               # Frequency we want the phase of
signal = np.sin(2*np.pi*f*t + np.pi/4)  # phase = -Ï€/4

def get_phase(signal, f):
    # FFT
    fft_vals = np.fft.fft(signal)
    freqs = np.fft.fftfreq(len(signal), 1/fs)

    # Find the index of frequency f
    idx = np.argmin(np.abs(freqs - f))

    # Phase at frequency f (radians)
    phase = np.angle(fft_vals[idx])
    return phase

print("Real phase  :", ((-np.pi/4)%(2*np.pi))*180/np.pi)
print("Phase (FFT) :", (get_phase(signal, f)%(2*np.pi))*180/np.pi)
```

```python
T = [10, 20, 200]
fs = 1

phases = np.zeros((3, max([p.shape[0] for p in planets])))
for i, planet in enumerate(planets):
    for j, signal in enumerate(planet):
        phase = get_phase(signal, 2*np.pi/T[i])
        phases[i, j] = phase

print(phases.mean(axis=1), phases.std(axis=1))
# std deviation is too high :(
```
